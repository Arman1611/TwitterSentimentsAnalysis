```python
import numpy as np ,pandas as pd
```

```python
data = pd.read_csv("0000000000002747_training_twitter_x_y_train.csv")
```

```python
data.head()
```

```python
data.tweet_location.fillna(' ',inplace=True)
```

```python
data.user_timezone.fillna(' ',inplace=True)
```

```python
data.airline.fillna(' ',inplace =True)
```

```python
text = list(data['text'])
airline = list(data['airline'])
loc = list(data['tweet_location'])
timezone = list(data['user_timezone'])
```

```python
result = list(data['airline_sentiment'])
```

```python
len(result)
```

```python
from nltk.corpus import stopwords
import string
stops = set(stopwords.words('english'))
punctuations = string.punctuation
stops.update(punctuations)
stops
```

```python
from nltk.stem import WordNetLemmatizer,wordnet
from nltk import pos_tag
from nltk.tokenize import word_tokenize
lemma = WordNetLemmatizer()
```

```python
def get_simple_pos(tag):
    if tag.startswith('J'):
        return 'a'
    elif tag.startswith('N'):
        return 'n'
    elif tag.startswith('V'):
        return 'v'
    elif tag.startswith('R'):
        return 'r'
    else:
        return wordnet.NOUN
```

```python
def clean_doc(text,a,l,t):
    text.extend([a,l,t])
    clean_words = []
    
    for w in text:
        if w.lower() not in stops:
            pos = pos_tag([w])
            word = lemma.lemmatize(w,pos = get_simple_pos(pos[0][1]))
            clean_words.append(word.lower())
            
    return clean_words
```

```python
x = []
for i in range(len(text)):
    words = clean_doc(word_tokenize(text[i]),airline[i],loc[i],timezone[i])
    x.append(words)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
x_train = [" ".join(doc) for doc in x]
```

```python
cv = CountVectorizer(max_features= 4105, stop_words=stops)
x_train_trans = cv.fit_transform(x_train)
x_train_trans.todense()
```

```python
cv.get_feature_names()
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
svc = MultinomialNB()
```

```python
svc.fit(x_train_trans,result)
```

```python
test = pd.read_csv('0000000000002747_test_twitter_x_test.csv')
```

```python
test.airline.fillna(' ',inplace =True)
test.tweet_location.fillna(' ',inplace =True)
test.user_timezone.fillna(' ',inplace =True)
```

```python
text = list(test['text'])
airline = list(test['airline'])
loc = list(test['tweet_location'])
timezone = list(test['user_timezone'])
```

```python
x_test = []
for i in range(len(text)):
    words = clean_doc(word_tokenize(text[i]),airline[i],loc[i],timezone[i])
    x_test.append(words)
```

```python
x_test = [" ".join(doc) for doc in x_test]
```

```python
x_test_trans = cv.transform(x_test)
```

```python
test_pred = svc.predict(x_test_trans)
```

```python
f = open('answer.csv','w')
for i in test_pred:
    f.write(i) 
    f.write('\n')
f.close()
```
